```python
# one sample t test
# one sample ttes is used to measure(to test hypothesis ) if the mean of the one group is the same as the certain number
from scipy import stats
import seaborn as sns
df=pd.read_csv("C://Users//cosmas//Desktop//machine learning project//Iris.csv")
df
```

```python
## to extract the sepal lengh data from data set
sepal_length_data=df["SepalLengthCm"]
sepal_length_data
```

```python
## to set the null hypothesis population mean 
hypothesized_mean=5.0
```

```python
# to perform one sample ttest
t_test,p_value=stats.ttest_1samp(sepal_length_data,hypothesized_mean)
t_test
```

```python
p_value
## since the p_value is greter than level of significance(5%) the we fail to reject the null hypothess,
```

```python
## to sample t test
# two sample ttest used to compare the means btn two groups 
from scipy import stats
import seaborn as sns
df=pd.read_csv("C://Users//cosmas//Desktop//machine learning project//Iris.csv")
df
```

```python
## to extarct the data 
sepal_length_setosa=df[df['Species']=='Iris-setosa']['SepalLengthCm']
sepal_length_versicolor=df[df['Species']=='Iris-versicolor']['SepalLengthCm']
```

```python
## to perfom the two sample ttest 
t_test,p_value=stats.ttest_ind(sepal_length_setosa,sepal_length_versicolor,equal_var=True)
p_value
```
